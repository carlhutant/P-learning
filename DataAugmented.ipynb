```python
import warnings
warnings.filterwarnings('ignore')
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

import os
import glob
import shutil
import pandas as pd
import numpy as np
from PIL import Image
import cv2
```

## Augumented 

```python
dataset = 'SUN'
folderType = 'val'

img_dir = './data/{}/IMG/{}'.format(dataset,folderType)
train_classes = pd.read_csv('./data/{}/trainclasses.txt'.format(dataset),header=None,sep = '\n')
test_classes = pd.read_csv('./data/{}/testclasses.txt'.format(dataset),header=None,sep = '\n')
```

```python
augment_to_num = 20
```

```python
def augmented_images(path,original_num, num, prefix):
    datagen = ImageDataGenerator(
                rotation_range=10,
                horizontal_flip=True,
                fill_mode='constant'
            )
    while( original_num < augment_to_num):
        images = glob.glob(path + '/*')

        rand_index = np.random.randint(len(images))
        img = load_img(images[rand_index])
        img = img_to_array(img)
        img = img.reshape((1,) + img.shape)
        
        for batch in datagen.flow(img, batch_size=1, save_to_dir=path, save_prefix=prefix, save_format='jpg'):
            original_num += 1
            if original_num % 10 == 0:
                print(original_num)
            break
```

## Train and Val 

```python
for class_name in train_classes[0]:
    img_path = os.path.join(img_dir,class_name)
    
    ## show the class image num 
    images = glob.glob(img_path + '/*')
    print("{}: {} Images".format(class_name, len(images)))
    original_num = len(images)
    augmented_images(img_path,original_num, augment_to_num, class_name)
```

## Test 

```python
for class_name in test_classes[0]:
    img_path = os.path.join(img_dir,class_name)
    
    ## show the class image num 
    images = glob.glob(img_path + '/*')
    print("{}: {} Images".format(class_name, len(images)))
    original_num = len(images)
    augmented_images(img_path,original_num, augment_to_num, class_name)
```
