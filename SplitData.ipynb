```python
import warnings
warnings.filterwarnings('ignore')

import os
import glob
import shutil
import pandas as pd
import numpy as np
from PIL import Image
import cv2
```

## split train , val , test

```python
### change the dataset here###
dataset = 'SUN'
##############################

train_classes = pd.read_csv('./data/{}/trainclasses.txt'.format(dataset),header=None,sep = '\n')
test_classes = pd.read_csv('./data/{}/testclasses.txt'.format(dataset),header=None,sep = '\n')
img_dir = './data/{}/IMG'.format(dataset)
```

## train and val

```python
for c in train_classes[0]:
    img_path = os.path.join(img_dir,c)
    images = glob.glob(img_path + '/*')
    print("{}: {} Images".format(c, len(images)))
    num_train = int(round(len(images) * 0.8))
    train, val = images[:num_train], images[num_train:]
    
    for t in train:
        if not os.path.exists(os.path.join(img_dir, 'train', c)):
            os.makedirs(os.path.join(img_dir,'train',c))
        shutil.move(t, os.path.join(img_dir,'train',c))
        
    for v in val:
        if not os.path.exists(os.path.join(img_dir, 'val', c)):
            os.makedirs(os.path.join(img_dir,'val',c))
        shutil.move(v, os.path.join(img_dir,'val',c))
```

## test

```python
for c in test_classes[0]:
    img_path = os.path.join(img_dir,c)
    images = glob.glob(img_path + '/*')
    
    for i in images:
        if not os.path.exists(os.path.join(img_dir, 'test', c)):
            os.makedirs(os.path.join(img_dir,'test',c))
        shutil.move(i, os.path.join(img_dir,'test',c))
        
```
